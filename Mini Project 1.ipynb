### 2. Plot a bar graph no. of matches won by each team in the year 2011

### Description of the problem
- import all the packages of matplotlib and numpy packages as we need for the required code.
- we take a sub database and store it differently separated under condition  year 2011
- we take the teams participated and number of wins of each team
- with the above data we create bar graph

```python
import matplotlib.pyplot as plt
import numpy as np
df2=df.copy()
sdf=pd.DataFrame(df2[df2['Season_Year'].isin(['2011'])])
wins=(sdf.groupby('match_winner')['Match_SK'].count())
teams=sdf.Team1.unique()
index=np.arange(len(teams))
plt.barh(teams,wins,color='Blue')
```
